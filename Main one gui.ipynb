```python
import tkinter as tk
from PIL import Image, ImageTk
import cv2
import numpy as np
import joblib
import pyscreenshot as ImageGrab
import os

# Initialize some variables
model = joblib.load("model/digit_recognizer")
lastx, lasty = None, None

def open_notebook():
    os.system('jupyter notebook MNIST_CLASSIFICATION.ipynb')

def prediction():
    img = ImageGrab.grab(bbox=(180,600,600,800))
    img.save("paint.png")
    im = cv2.imread("paint.png")
    load = Image.open("paint.png")
    load = load.resize((280,280))
    photo = ImageTk.PhotoImage(load)

    img = tk.Label(canvas3,image=photo,width=280,height=280)
    img.image=photo
    img.place(x=0,y=0)

    im_gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    im_gray = cv2.GaussianBlur(im_gray,(15,15),0)
    ret,im_th = cv2.threshold(im_gray,100,255,cv2.THRESH_BINARY)
    roi = cv2.resize(im_th,(28,28),interpolation=cv2.INTER_AREA)

    rows,cols=roi.shape
    X = []
    for i in range(rows):
        for j in range(cols):
            k = roi[i,j]
            if k>100:
                k=1
            else:
                k=0
            X.append(k)
    #x = np.array(x).reshape(1, -1)  # Reshape the input for prediction
    prediction = model.predict([X])

    result_label.config(text="Prediction = {}".format(prediction[0]))
    a1 = tk.Label(canvas3,text="Prediction = ",font=("Algerian,20"))
    a1.place(x=5,y=550)
    b1 = tk.Label(canvas3,text=prediction[0],font=("Algerian,20"))
    b1.place(x=200,y=550)

def activate_paint(e):
    global lastx, lasty
    canvas2.bind('<B1-Motion>',paint)
    lastx, lasty = e.x,e.y

def paint(e):
    global lastx,lasty
    x,y = e.x,e.y
    canvas2.create_line((lastx,lasty,x,y),width=20,fill="white")
    lastx,lasty = x,y

def clear():
    canvas2.delete("all")

# Initialize the main window
window = tk.Tk()
window.title("Handwritten Digit Classification")
window.geometry("800x600")
window.resizable(0, 0)

# Initialize some widgets
header = tk.Button(window,text="HANDWRITTEN DIGIT CLASSIFICATION",font=('Algerian',30),bg="green",fg="black")
header.pack()

canvas1 = tk.Canvas(window, width=500, height=250, bg="ivory")
canvas1.place(x=5, y=120)

canvas2 = tk.Canvas(window,width=500,height=270,bg="black")
canvas2.place(x=5,y=380)

canvas3 = tk.Canvas(window,width=280,height=530,bg="green")
canvas3.place(x=515,y=120)

b1 = tk.Button(canvas1,text="MNIST DATASET MODEL CLASSIFICATION",font=('Algerian',15),bg="orange",fg="black", command=open_notebook)
b1.place(x=10,y=50)

b2 = tk.Button(canvas1,text="LIVE PREDICTION",font=('Algerian',15),bg="pink",fg="blue",command=prediction)
b2.place(x=10,y=120)

# b3 = tk.Button(canvas1, text="CALCULATING ACCURACY OF BOTH THE MODELS", font=('Algerian', 15), bg="green", fg="white", command=lambda: show_image("accuracy.png", canvas3))
# b3.place(x=10,y=140)

b4 = tk.Button(canvas1, text="EXIT", font=('Algerian', 15), bg="red", fg="white", command=window.destroy)
b4.place(x=10, y=200)

result_label = tk.Label(canvas3, text="", font=('Algerian', 20),bg="pink")
result_label.place(x=5, y=350)

canvas2.bind('<1>',activate_paint)

btn = tk.Button(canvas2,text="clear",fg="white",bg="red",command=clear)
btn.place(x=0,y=0)

# Initialize the main event loop
window.mainloop()
```
